```python
import os
import shutil
  
source = 'D:/ISIC_2019_Training_Input/ISIC_2019_Training_Input/'
destination = 'D:/ISIC_2019_Training_Input/Melanoma/'
  
allfiles = os.listdir(source)
  
for f in allfiles:
    m=f.split('.')[0]
    if m in mel_list:
        file="melanoma_" + str(f)
        print(file)
        shutil.copy(source + f, destination + file)
        
#     file="thorsten_Neutral_" + str(f)
#     shutil.copy(source + f, destination + file)
 #   print(f)
```
